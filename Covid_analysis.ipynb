```python
def first_digit_distribution(dataset):
    for x in range(len(dataset)):
        while dataset[x] >= 10:
            dataset[x]= dataset[x]/10
    first_digits=[int(x) for x in sorted(dataset)]
   
    data_count=[]
    for i in range(1,10):
        count=first_digits.count(i)
        data_count.append(count)
    total_count=sum(data_count)
    data_distribution=[100* (n/total_count) for n in data_count]
    return data_distribution
```

```python
first_digit_distribution(cumulativecases[20:])
```
